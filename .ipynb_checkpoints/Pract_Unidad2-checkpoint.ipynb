3) Calcule el promedio de lo facturado en todos los vuelos realizados considerando únicamente los registros de viajes que se facturaron en USD

```python
facturacion = cleaned_data[['vuelo','factura_moneda_monto']].dropna(how='all')
```

```python
facturacion["moneda"] = facturacion["factura_moneda_monto"].str.split().str[0] # Se separa el tipo de moneda en columna aparte
```

```python
facturacion = facturacion[["vuelo", "moneda", "factura_moneda_monto"]] # Se reordena el dataset con datos relevantes
```

```python
facturacion["factura_moneda_monto"] = facturacion["factura_moneda_monto"].str.replace(r"USD |EUR ", "", regex=True)
```

```python
facturacion["factura_moneda_monto"] = (
    facturacion["factura_moneda_monto"]
    .str.replace(".", "", regex=False)  # Se eliminan puntos
    .str.replace(",", ".", regex=False)  # Coma por punto
    .str.replace(" ", "", regex=False)  # Se borran espacios en blanco
    .astype(float)  # Datos a float
)
```

```python
facturacion_usd = facturacion[facturacion["moneda"] == "USD"]
facturacion_usd["factura_moneda_monto"].mean().round(2)
```

### Ejercicio N°2

```python
raw_incendios = pd.read_csv(r'C:\Users\icrui\Documents\DS-ML\FCEIA\3° cuatrimestre\Fundamentos_Cs_Datos\Unidad2\incendios-cantidad-causas-provincia_2022.csv', header=3, encoding='latin1')
```

```python
raw_incendios['provincia'] = raw_incendios['provincia'].str.strip()
```

```python
raw_incendios['provincia'].value_counts()
```

```python
raw_incendios.isna().sum()
```

```python
cleaned_incendios = raw_incendios
```

1) Obtenga el número de incendios totales por año para todo el país. ¿Cuál fue el año en el que se presentó un mayor número de incendios?

```python
incendios_totales = cleaned_incendios.groupby(['provincia', 'anio'])['total'].sum().reset_index().sort_values(by='total', ascending=False)
incendios_totales
```

2) Obtenga el número de incendios totales por año para el período 1993-2021 en la provincia de Córdoba.

```python
incendios_totales[(incendios_totales['provincia'] == 'Córdoba') 
    & (incendios_totales['anio'] >= 1993) 
    & (incendios_totales['anio']<= 2021)].sort_values(by='anio', ascending=False)
```

3) Realice una tabla en la que se muestre, para cada año del periodo 1993-2021, la provincia en la que tuvo lugar el mayor número de incendios intencionales. Sugerencia: explore las funcionalidades del método idxmax() de la librería Pandas.

```python
incendios_intencionales = cleaned_incendios[(cleaned_incendios['anio'] >= 1993) 
    & (cleaned_incendios['anio'] <=2021) 
    & (cleaned_incendios['intencional'] > 0)].groupby(['provincia', 'anio'])['intencional'].sum().reset_index()
idx_max = incendios_intencionales.groupby('anio')['intencional'].idxmax()
tabla_incendios_intencionales = incendios_intencionales.loc[idx_max].sort_values(by='anio')
tabla_incendios_intencionales
```

4) Realice un gráfico de barras para visualizar el número de incendios intencionales, por negligencia y naturales que tuvieron lugar durante el periodo 2015-2021 en la provincia de Santa Fe.

```python
incendios_stafe = cleaned_incendios[(cleaned_incendios['provincia']=='Santa Fe') 
    & (cleaned_incendios['anio'] >= 2015)
    & (cleaned_incendios['anio'] <= 2021)]
cleaned_stafe = incendios_stafe[['anio','intencional', 'negligencia', 'natural']]
cleaned_stafe
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker

anios = cleaned_stafe['anio']
tipos_incendio = ['intencional', 'negligencia', 'natural']
x = np.arange(len(anios))  # posición de los grupos (años)
width = 0.25  # ancho de cada barra

fig, ax = plt.subplots(figsize=(10, 4))

# Dibujar barras y guardar objetos para etiquetar
for i, tipo in enumerate(tipos_incendio):
    barras = ax.bar(x + i * width, cleaned_stafe[tipo], width, label=tipo.capitalize())

    # Agregar etiquetas sobre cada barra
    for bar in barras:
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2, 
            height + 1,  # +1 para dejar un pequeño espacio
            f'{int(height)}', 
            ha='center', va='bottom', fontsize=8
        )

# Estética
ax.set_xlabel('Año')
ax.set_ylabel('Cantidad de incendios')
ax.set_title('Cantidad de incendios por tipo y año')
ax.set_xticks(x + width)
ax.set_xticklabels(anios, rotation=45)
ax.legend()
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```
